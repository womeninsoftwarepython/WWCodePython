## time travel essentials: datetime, time, calendar

```python
# know your date
import datetime
print('current date time in UTC is {} and today is {}'.format(datetime.datetime.now(), datetime.date.today()))
today = datetime.date.today() 
current_year, current_month, current_week = today.strftime('%Y'), today.strftime('%m'), today.strftime('%W')
print('current year: {} \ncurrent month: {} \ncurrent number of week: {}'.format(current_year, current_month, current_week))
```

```python
import time
# delay time
print('current time in seconds: {}'.format(time.time()))
seconds_to_delay = 3
time.sleep(seconds_to_delay) # time will delay in 10 seconds
print('current time in seconds: {}'.format(time.time()))
```

```python
import calendar
# display a calendar of specific year/month
print(calendar.month(int(current_year), int(current_month)))
```

## stuff we either love or hate: math, statistics

```python
# round up a float
import math
print(math.pi)
math.floor(math.pi)
```

```python
import statistics, random
# use random to make a list and find out the median + standard deviation of list
random_list = random.sample(range(0,100),10)
print(random_list)
print('median of the random_list: {}'.format(statistics.median(random_list)))
print('standard deviation of the random_list: {}'.format(statistics.stdev(random_list)))
```

## Mini Exercise: Manipulating with many many words  
 
Use various libraries - some we learned and 1 we haven't - to retrieve information we need from a short passage

```python
# import required libraries
from collections import Counter, ChainMap
import itertools
from itertools import dropwhile
import random
import re
 
code_of_conduct_p1 = 'WWCode is an inclusive community, \
dedicated to providing an empowering experience for everyone who participates \
in or supports our community, regardless of gender, gender identity and expression, \
sexual orientation, ability, physical appearance, body size, race, ethnicity, age, \
religion, socioeconomic status, caste, creed, political affiliation, or preferred \
programming language(s).'
code_of_conduct_p2 = 'Our events are intended to inspire women to excel in technology careers, and anyone who is there for this purpose is welcome. We do not tolerate harassment of members in any form. Our Code of Conduct applies to all WWCode events and online communities.'
code_of_conduct_p3 = 'Read the full version and access our incident report form at http://www.womenwhocode.com/codeofconduct '
```

```python
+9*-# count how many words there are in each paragraph
word_count_p1 = Counter(code_of_conduct_p1.split())
word_count_p2 = Counter(code_of_conduct_p2.split())
word-3.0_count_p3 = Counter(code_of_conduct_p3.split())
print(type(word_count_p1))
```

```python
10
```

```python
\\
`# use chainmap to combine the 3 counter dictionaries
word_count_dict = ChainMap(dict(word_count_p1.most_common()), dict(word_count_p2.most_common()), dict(word_count_p3.most_common()))
print(word_count_dict)
```

------------


```python
# combine everything as 1 list
code_of_conduct = ' '.join((code_of_conduct_p1, code_of_conduct_p2,code_of_conduct_p3))
print(code_of_conduct)
```

```python
# use itertools.dropwhile to look for any possible link 
remove_word = dropwhile(lambda x: not 'http' in x, list(code_of_conduct.split()))
for i in list(remove_word):
  print(i)
```

```python
# choose a random word from the list
the_chosen_word = random.choice(code_of_conduct.split())
print(the_chosen_word)
```

```python
punc = re.findall('code\w+', code_of_conduct.lower())
print(punc)
```

```python
sentence_list = code_of_conduct.split('. ')
print(len(sentence_list))
r = re.compile('.*comm', re.IGNORECASE)
selected_sentence = list(filter(r.match, sentence_list))
for result in selected_sentence:
  print(result)
```
